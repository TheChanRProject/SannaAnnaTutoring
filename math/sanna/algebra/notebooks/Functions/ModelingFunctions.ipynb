```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

xVals = np.array([-1,1,2,4,5])
yVals = np.array([0,-4,-3,5,12])

def quadratic(x):
    func = (x-1)**2 - 4
    return func 

yVals2 = quadratic(xVals)
    

# Plot

plt.plot(xVals,yVals,'r')
plt.plot(xVals, yVals2, 'g')
plt.show()
```

```python
xRay = np.arange(0,5)
yRay = np.array([1,4,5,4,1])

# Function 

def quad2(x):
    return -(x-2)**2 + 5

yRay2 = quad2(xRay)
# Plot


plt.plot(xRay, yRay, 'b')
plt.plot(xRay,yRay2,'c')
plt.show()
```
