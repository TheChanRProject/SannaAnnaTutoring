Describe the horizontal and vertical translations of TRI to T'R'I' .

```python
%matplotlib inline
import math 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
preXVals = np.array([-3, 2, 2, -3])
preYVals = np.array([2, 0, 2, 2])

plt.plot(preXVals, preYVals, 'r')
plt.show()
```

Let's look at the coordinates of R and R' .

R is situated at (2,2) and R' is situated at (-5,-3) . 

The translation from R to R' can be mapped using the following formula: 

$$ (x_R + m , y_R + n) \rightarrow (x_{R'}, y_{R'}) $$ 

m represents the horizontal change in the translation and n represents the vertical change in the translation. 

Using this model, we can now solve for both m and n after plugging in the values of R and R' .

$$ (2 + m, 2 + n) \rightarrow (-5, -3) $$
$$ 2 + m = -5 $$ 
$$ 2 + n = -3 $$ 
$$ m = -5 - 2 \rightarrow -7 $$
$$ n = -3 - 2 \rightarrow -5 $$  

From these calculations, we can see that the translation from TRI to T'R'I' has a horizontal change of -7 and a vertical change of -5. 

```python
postXVals = preXVals - 7
postYVals = preYVals - 3 

```

```python
plt.plot(preXVals,preYVals, 'r')
plt.plot(postXVals,postYVals, 'g')
plt.show()
```
