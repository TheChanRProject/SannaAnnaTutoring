We can see the transformation visually from $ f(x) = 3^x $ to $ 3^{x-1} - 5 $ 

There is no reflection so that's why the graph moves to the right by one unit and down by 5 units. 

We can see a graphical presentation of this.

```python
def absolute(a,x,h,k):
    return a*abs(x-h) + k

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt 

xVals = np.arange(-4,5)
func1Vals = absolute(1,xVals,0,0)
func2Vals = absolute(1,xVals,1,-5)

plt.plot(xVals,func1Vals, 'r')
plt.plot(xVals, func2Vals, 'g')
plt.show()
```
