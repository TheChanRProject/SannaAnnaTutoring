```python
def double(x):
    return 2*x

import numpy as np

xArray = np.array(range(11))
print(xArray)
yArray = double(xArray)
print(yArray)

%matplotlib inline

import matplotlib.pyplot as plt

plt.plot(xArray,yArray, 'r-')
plt.show()
```

```python
xList = np.array(range(11))
yList = np.array(range(11))

def rotate(a, b: np.array, c: np.array):
    if a == 90:
        return -c, b
    if a == 180:
        return -b, -c
    if a == 270:
        return c, -b

apple = rotate(270,xList,yList)
banana = rotate(90, xList, yList)
pear = rotate(180, xList, yList)
```

```python
plt.plot(xList,yList, 'ro')
plt.plot(apple[0], apple[1], 'go')
plt.plot(banana[0], banana[1], 'yo')
plt.plot(pear[0], pear[1], 'bo')
plt.show()
```
