The parent function of the given transformation is $$ y = \sqrt{x} $$ The transformations that were applied to the function are as follows:

There is a horizontal stretch of 2, the graph moved 6 units left, and 2 units downwards.

$$ y = \sqrt{x + 6} - 2 $$ 

Let's view it graphically

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt 

def rootFunction(a,x,h,k):
    return np.sqrt(a*x - h) + k

xVals = np.arange(-10,11, 0.1)
Y1Vals = rootFunction(1,xVals,0,0)
Y2Vals = rootFunction(2,xVals,-6,-2)

plt.plot(xVals,Y1Vals,'r')
plt.plot(xVals,Y2Vals, 'g')
plt.show()
```
